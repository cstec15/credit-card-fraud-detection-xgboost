Before going any further on trying to fine tune our logistic regression model, we will do a baseline check of a random forest. Random forests can be advantageous as they can capture nonlinear interactions between features, are less sensitive to outliers, and can perform better on imbalanced data. It may be less interpretable as there are no coefficients like in logistic regression, but since we are already dealing with principal components, much of the interpretability is already gone. 

## Random Forest

```python
import pandas as pd
import numpy as np

creditcards = pd.read_csv("~/credit-card-fraud-detection-xgboost/data/creditcards_cleaned.csv")
correlations = creditcards.corr()['Class'].drop('Class')
most_correlated = correlations.abs().sort_values(ascending=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, average_precision_score, precision_recall_curve
import matplotlib.pyplot as plt
import seaborn as sns

X = creditcards.drop('Class', axis=1)[list(most_correlated.index[:5])]
y = creditcards['Class']

# Train-test split (with stratify to preserve class balance)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# n_estimators is number of decision trees used, class_weight balanced makes fraud samples heavier in the loss function
rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)

# Predictions and probabilities
y_pred = rf_model.predict(X_test)
y_proba = rf_model.predict_proba(X_test)[:, 1]

# Confusion matrix & classification report
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# AUPRC score
auprc = average_precision_score(y_test, y_proba)
print(f"AUPRC Score: {auprc:.3f}")

# Plot Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label=f'AUPRC = {auprc:.3f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

With no additional tuning, the model performed significantly better in our baseline measure of success (AUPRC) with a score of .812 and our best Logistic Regression model so far having a score of about .70. Most of this increase can be attributed to an increase in precision, with the amount of false positives decreasing significantly. Due to this increased performance, we will continue on tuning our random forest model as our prime candidate. There is still room for improvement, specifically in the false negative category, as we may view missing true fraud as more harmful than false positives. 

We will now try different prediction thresholds in our random forest to try to decrease the false negative rate

```python
from sklearn.metrics import fbeta_score

thresholds = np.arange(0, 1.01, 0.01)
f2_scores = []

for thresh in thresholds:
    y_pred = (y_proba >= thresh).astype(int)
    f2 = fbeta_score(y_test, y_pred, beta=2)
    f2_scores.append(f2)

# Find the threshold with the highest F2
best_thresh_idx = np.argmax(f2_scores)
best_thresh = thresholds[best_thresh_idx]
best_f2 = f2_scores[best_thresh_idx]

print(f"Best threshold for F2: {best_thresh:.2f}")
print(f"F2 Score at this threshold: {best_f2:.3f}")
```

It does not appear that adjusting the threshold alone will allow us to decrease our false negatives without drastically increasing our false positives.

Next, we will use Grid Search to find the model parameters that give us the best performance


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, average_precision_score

X = creditcards.drop('Class', axis=1)[list(most_correlated.index[:5])]
y = creditcards['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# define parameters to test 
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt'],
    'class_weight': ['balanced']
}

```

```python
rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='average_precision',
    n_jobs=1,
    verbose=1
)

grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("AUPRC Score:", average_precision_score(y_test, y_proba))
```

We can see that our best results from Grid Search did not perform any better than our original parameters. I will now try with more features included and utilize RandomizedSearchCV for faster performance with almost as good of results as GridSearch without searching every possible combination

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix, average_precision_score

# Split your data
X = creditcards.drop('Class', axis=1)[list(most_correlated.index[:15])]
y = creditcards['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

param_dist = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [None, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'class_weight': ['balanced']
}
```

```python
rf = RandomForestClassifier(random_state=42)

random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=10,
    scoring='average_precision',
    cv=3,
    random_state=42,
    n_jobs=1,
    verbose=1
)

random_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters:", random_search.best_params_)

best_rf = random_search.best_estimator_

# Evaluate on test set
y_pred = best_rf.predict(X_test)
y_proba = best_rf.predict_proba(X_test)[:, 1]

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("AUPRC Score:", average_precision_score(y_test, y_proba))
```

Similar results with random search. Even after testing various hyperparameters, AUPRC did not increase.
