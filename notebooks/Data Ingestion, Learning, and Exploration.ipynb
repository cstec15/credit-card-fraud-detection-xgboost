### Let's start with visualizing a few variables and relationships to understand our dataset

I want to explore if there are certain times of the day that are more likley to have fraud.

```python
creditcards['Hour of Day'] = (creditcards['Time'] // 3600 % 24).astype(int)
fraud_ratio = creditcards.groupby('Hour of Day')['Class'].mean()
plt.figure(figsize=(12,6))
sns.barplot(x=fraud_ratio.index, y=fraud_ratio.values, palette="mako")
plt.title('Proportion of Fraud Transactions per Hour')
plt.xlabel('Hour of Day (Relative to Start of Dataset)')
plt.ylabel('Proportion of Fraudulent Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fraud_count = creditcards[creditcards['Class'] == 1].groupby('Hour of Day').size()

plt.figure(figsize=(12,6))
sns.barplot(x=fraud_count.index, y=fraud_count.values, palette="rocket")
plt.title('Total Number of Fraud Transactions per Hour')
plt.xlabel('Hour of Day (Relative to Start of Dataset)')
plt.ylabel('Count of Fraudulent Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

While we do not know what time of day the dataset begins at, we can look at times relative to the start of the dataset. If we set start time of the dataset at 0000, it appears that a higher proportion of transactions are fraud between 0200 and 0400, and particularly at 0200. When looking purely at count of fraudulent detections, 0200 still sticks out, but 1100 also has a much higher count relative to the other hours.

There appears to be different rates/likelihood of fraud at different times of day, so we will create a dummy variable for hour that will allow us to take this into account in any models.

```python
creditcards = pd.get_dummies(creditcards, columns=['Hour of Day'], drop_first=True)
creditcards.head()
```

We will standardize the Amount column to help with model stabiity and interpretability

```python
creditcards['Amount_standardized'] = (creditcards['Amount'] - creditcards['Amount'].mean()) / creditcards['Amount'].std()
creditcards.drop('Amount', axis=1, inplace=True)
```

Next we will plot a correlation matrix to understand the relationships between variables

```python
plt.figure(figsize=(15, 10))
sns.heatmap(creditcards.corr(), cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Matrix of All Features')
plt.show()
```

```python
corr_matrix = creditcards.corr().abs()
mask = np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
corr_matrix_filtered = corr_matrix.where(mask)
sorted_corr = corr_matrix_filtered.unstack().dropna().sort_values(ascending=False)
top_n = 20
print(sorted_corr.head(top_n))
correlations = creditcards.corr()['Class'].drop('Class')
most_correlated = correlations.abs().sort_values(ascending=False)
```

The variables V17, V14, V12, and V10 are all correlated with Class, so we will look to include these in any parametric models we choose to use as they may be predicitive.

We will further investigate the relationship between Amount and V2 to look for multicolinearity 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

features = ['Amount_standardized', 'V2', 'V7', 'V5', 'V20', 'V1', 'V6', 'V3']
X = add_constant(creditcards[features])
vif = pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index=X.columns)
print(vif)
```

This is not an extreme VIF, but we will keep an eye on it if our model is not performing

```python
cols_to_plot = [
    col for col in creditcards.columns 
    if col not in ['Time', 'Class'] and not col.startswith('Hour of Day_')
]
cols = 4
rows = (len(cols_to_plot) + cols - 1) // cols

plt.figure(figsize=(20, rows * 4))

for i, col in enumerate(cols_to_plot):
    plt.subplot(rows, cols, i + 1)
    creditcards.boxplot(column=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

It appears that many of the features are very skewed and/or have many outliers. We will fit an initial logistic regression model on the features most correlated with Amount to get a baseline and iteratively improve on it from there.

We will use AUPRC as our performance metric because of the rarity of the fraud class.
